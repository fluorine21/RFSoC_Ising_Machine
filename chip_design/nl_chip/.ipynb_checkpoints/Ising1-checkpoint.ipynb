```python
from phidl import Path, CrossSection, Device
from phidl import quickplot as qp # Rename "quickplot()" to the easier "qp()"
import phidl.geometry as pg
import phidl.utilities as pu
import phidl.path as pp
import layoutISING as li
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import pi, c
```

# Initialization

Load process layers and save layer properties file for Klayout visualization.

```python
ls = li.setup_layers()
P = pg.preview_layerset(ls)
qp(P)
# P.write_svg('Layers.svg')
pu.write_lyp('Ising0.lyp', layerset = ls)
```

Units

```python
mm = 10**3
um = 1 
nm = 10 ** (-3)
```

```python
# Global Chip Parameters
chip_width = 23*mm
chip_height = 13*mm
T = li.chip(size=(chip_width,chip_height), name='Ising0', text_size=150, keepout=1.5*mm)

X0 = 0
Y0 = 2300
ΔX = chip_width
ΔY = 1048

height = ΔY
for k in range(9):
    T << pg.rectangle( size=(chip_width, height), layer=98 ).move( [0,Y0+k*ΔY-height/2] )

qp(T)
```

Pads

```python
D = Device('Ising')
D << pg.rectangle( size=(chip_width, height), layer=0 )

# Bonding Pads
pad = 50*um;
pad_g = 100*um;
elec = pad;
elec_m = 14*um; #middle_e_width
elec_s = 28*um; #side_e_width
pad_m = 1.5*mm

#RF pads
n_rf = 4 #number of RF pobres
spacing = 300
xoff = chip_width - (n_rf*2*spacing-pad) - pad_m 
yoff = height-pad #-pad/2

rf_x = xoff
rf_y = yoff
for k in range(4*2):
    D << li.triplet_pad(xoff+k*spacing,yoff,pad,pad_g,elec_m,elec,elec_s,e_e_gap= 9*um,pos='u')
    
#DC pads
n_dc = 16 #number of DCpobres
g_dc = 100
xoff = 2*mm #(chip_width - (n_dc*g_dc-pad))/2
yoff = pad/2
for k in range(n_dc):
    D << li.dc_pad(xoff+k*g_dc,yoff,pad,str(k+1))
 
#Label 
L = pg.text(text='CIM_V0-1', size=40, layer=30, justify='center')
D.add_ref(L).move((1.8*mm,yoff))
D.write_gds('Ising0.gds')
```

QPM Hunt Wgs

```python
wg_width = 1.8*um
wg_width_oc = 2.3*um #outside cavity
wg_width_ic = 2.5*um #inside cavity
wg_width_m2m = 2.2*um #main to memory wg  NEED TO ADJUST THE RF EOM ANGLES FOR SPECIFIC CASE
wg_width_nif = 1.9*um
V_Groove_Spacing = 127*um;
xmargin =0.5*mm

wg_y_off = (height-V_Groove_Spacing*8)/2

rVg = 100*um
anglerVg = li.find_ey(V_Groove_Spacing/2,radius=rVg)
rVg_euler_x =li.mod_euler(radius = rVg, angle = anglerVg)[1][0] 

#bottom wg QPM Main
P = Path()
xbb = 3500

P.append( pp.straight(length = xbb+xmargin) )  
P.append( pp.euler(radius = rVg, angle = -anglerVg) )
P.append( pp.euler(radius = rVg, angle = anglerVg) )
P.append( pp.straight(length = chip_width-xbb+xmargin) ) 
P.movex(-xmargin)
P.movey(wg_y_off+V_Groove_Spacing)

X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(X)

#top wg QPM Memory
P = Path()
R = V_Groove_Spacing/4
xbt = 3000
s_angle = 30
clearance = 200*um

sl1 = chip_width - (n_rf*2*spacing+pad*4) - pad_m   - clearance
P = Path()
P.append( pp.straight(length = sl1+xmargin) )
P.append( pp.euler(radius = rVg, angle = -anglerVg) )
P.append( pp.euler(radius = rVg, angle = anglerVg) )
P.append( pp.straight(length = chip_width-sl1+xmargin) )
P.movex(-xmargin)
P.movey(wg_y_off+V_Groove_Spacing*8)

X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(X)
D.write_gds('Ising0.gds')
```

EOM DC 1 (Pump IM)

```python
im_gap = 0.8*um
im_length = 1.5*mm
coupler_l1 = 100*um
im_r = 50*um
im_angle = 45
elec_w = 50*um
e_e_gap = 10*um
via = elec_w*0.8

x_pos = 1.5*mm
y_pos = wg_y_off+4*V_Groove_Spacing

IM = Device('IM')
(IM,xl) = li.dcim(im_gap, im_length, coupler_l1, im_r, im_angle,elec_w,e_e_gap,via,wg_width_oc,V_Groove_Spacing)
D << IM.move([x_pos, y_pos])

#straight input wg top
P = Path()
P.append( pp.straight(length = x_pos+xmargin) )
P.movex(-xmargin)
P.movey(y_pos+V_Groove_Spacing)

X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

#straight input wg bottom
P = Path()
P.append( pp.straight(length = x_pos+xmargin) )
P.movex(-xmargin)
P.movey(y_pos)

X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)
D.write_gds('Ising0.gds')

eomdc_1xs = x_pos
eomdc_1xl = xl*2+im_length
eomdc_1yb = y_pos
eomdc_1yt = y_pos+V_Groove_Spacing
```

```python
# MMI Parameters
W = 2.3*um
L_tp = 5*um
W_tp = 2*um
L_mmi = 40*um
W_mmi = 8*um
Y_mmi = 4*um

# MZI Parameters
radius = 100
angle = 12.814353309

# Electrode Parameters
middle_e_width = 14*um
side_electrode_width = middle_e_width*2
e_e_gap = 9*um

# Heater Parameters
w_heater = 5*um
g_heater = 3*um
wg_gap = w_heater+2*g_heater+wg_width_oc
off_chip = 0

# Inter Cavity Bend Parameters
rt_start = 3900*um

interR = 70*um
interA = 45
inter_euler_x =li.mod_euler(radius = interR, angle = interA)[1][0]
inter_euler_y =li.mod_euler(radius = interR, angle = interA)[1][1]
coupler_g1 = 1*um #coupler bottom/top
coupler_g2 = 1*um #coupler middle
coupler_g3 = 1*um #coupler interferomter
rt_r = (height/2 - 2*inter_euler_y - wg_gap - wg_y_off-wg_width_oc/2-wg_width_m2m/2 -wg_width_ic - coupler_g1*2)/2
rt_rt = rt_r+(wg_width_ic-wg_width_m2m)/4

coupler_lp = 90*um
coupler_lall = 110*um
rfeom_elecR = 50*um;
```

EOM DC 2 (Main Cavity Pump IM)

```python
length = 1.5*mm
x_pos = 0; y_pos = 0 #dummy values for now
xl = li.rfim(wg_width_oc, off_chip, W, L_tp, W_tp, L_mmi, W_mmi,Y_mmi,length, radius,x_pos,y_pos,middle_e_width,e_e_gap)[1]

x_pos = eomdc_1xs + (eomdc_1xl - xl*2-length)/2
y_spacing = ((wg_y_off+V_Groove_Spacing*8) - wg_gap - eomdc_1yt)/2
y_pos = eomdc_1yt + y_spacing

#eom
DC2 = Device('DC2')
DC2 = li.dcimthin(wg_width_oc, off_chip, W, L_tp, W_tp, L_mmi, W_mmi,Y_mmi,length, radius,x_pos,y_pos,middle_e_width,e_e_gap)[0]
D << DC2

#wg connection from eom 1
P = Path()
P.append( pp.arc(radius = y_spacing/2, angle = 180) )
P.append( pp.straight(length = (eomdc_1xl - xl*2-length)/2) )
P.movex(eomdc_1xs+eomdc_1xl)
P.movey(eomdc_1yt)

X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

#wg connection from eom -> straight wg
uptocoupler2 = rt_start + rt_r - 4*rVg_euler_x - length - x_pos
P = Path()
P.append( pp.arc(radius = y_spacing/2, angle = -180) )
P.append( pp.straight(length = length ))
P.append( pp.euler(radius = rVg, angle = -anglerVg) )
P.append( pp.euler(radius = rVg, angle = anglerVg) )
P.append( pp.euler(radius = rVg, angle = anglerVg) )
P.append( pp.euler(radius = rVg, angle = -anglerVg) )
P.append( pp.straight(length = uptocoupler2 ))
P.movex(x_pos)
P.movey(y_pos)
P.rotate(180, center = [x_pos,y_pos])
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

Ramp = pg.ramp(length = coupler_l1, width1 = wg_width_oc, width2 = wg_width_ic, layer = 30)
Ramp.mirror(p1 = [0,0], p2 = [1,0])
D << Ramp.move([rt_start+rt_r,wg_y_off+V_Groove_Spacing*8 -wg_gap+wg_width_oc/2])

D.write_gds('Ising0.gds')
```

Heater Top

```python
ht_sx = x_pos+length+rVg_euler_x*3/2
ht_sy = wg_y_off+V_Groove_Spacing*8-g_heater-w_heater-wg_width_oc/2
ht_ex = ht_sx+sl1-ht_sx-pad
D << pg.rectangle( size=(sl1-ht_sx, w_heater), layer=60 ).move( [ht_sx,ht_sy] )
D << pg.rectangle( size=(pad,pad-7.5), layer=60 ).move( [ht_sx,ht_sy-pad+7.5] )
D << pg.rectangle( size=(pad,2*pad-22.5), layer=60 ).move( [ht_ex,ht_sy-2*pad+22.5] )
D.write_gds('Ising0.gds')
```

EOM DC 3 (Memory Cavity Pump PM)

```python
length = 1.5*mm

x_pos = eomdc_1xs + (eomdc_1xl-length)/2
y_pos = eomdc_1yb - y_spacing

#eom
DC3 = Device('DC3')
DC3 = li.dcpm(length,elec_w,e_e_gap,via,wg_width_oc)
D << DC3.move([x_pos,y_pos])

#wg connection from eom 1
P = Path()
P.append( pp.arc(radius = y_spacing/2, angle = -180) )
P.append( pp.straight(length = (eomdc_1xl -length)/2) )  #adjust length later to match pump length
P.movex(eomdc_1xs+eomdc_1xl)
P.movey(eomdc_1yb)

X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

#wg connection from eom -> straight wg
bend_offset = 200*um
sb = xbb-x_pos+bend_offset

uptocoupler = rt_start+rt_r-x_pos-sb-2*rVg_euler_x

P = Path()
P.append( pp.arc(radius = y_spacing/2, angle = 180) )
P.append( pp.straight(length = sb ))  
P.append( pp.euler(radius = rVg, angle = -anglerVg) )
P.append( pp.euler(radius = rVg, angle = anglerVg) )
P.append( pp.straight(length =uptocoupler) ) 
P.movex(x_pos)
P.movey(y_pos)
P.rotate(180, center = [x_pos,y_pos])
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

Ramp = pg.ramp(length = coupler_l1, width1 = wg_width_oc, width2 = wg_width_ic, layer = 30)
D << Ramp.move([rt_start+rt_r,wg_gap+wg_y_off-wg_width_oc/2])

D.write_gds('Ising0.gds')
```

EOM RF0 (Memory->Main Cavity PM)

```python
length0 = 4*mm
x_pos0 = rt_start+rt_r+coupler_lall+inter_euler_x*2
y_pos0 = height/2
D << li.rfpm(wg_width_ic, length0, middle_e_width, e_e_gap).move([x_pos0, y_pos0])

P = Path()
P.append( pp.euler(radius = interR, angle = -interA) )
P.append( pp.euler(radius = interR, angle = interA) )
P.movex(x_pos0)
P.movey(y_pos0)
P.rotate(180, center = [x_pos0,y_pos0])
X = CrossSection()
X.add(width = wg_width_ic, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

Ramp = pg.ramp(length = coupler_lall, width1 = wg_width_m2m, width2 = wg_width_ic, layer = 30)
D << Ramp.move([x_pos0-coupler_lall-inter_euler_x*2,y_pos0+2*inter_euler_y-wg_width_ic/2])

D << li.fang(wg_width_m2m,length=100*um,orientation='LD').move([x_pos0-coupler_lall-inter_euler_x*2,y_pos0+2*inter_euler_y-(wg_width_ic-wg_width_m2m)/2])

D.write_gds('Ising0.gds')
```

EOM RF1 (Memory->Main Cavity IM)

```python
length1 = 4*mm
x_pos1 = x_pos0+length0+rfeom_elecR*2
y_pos1 = height/2
D << li.eom23(wg_width_m2m, off_chip, W, L_tp, W_tp, L_mmi, W_mmi,Y_mmi,length1, radius,x_pos1,y_pos1,middle_e_width,e_e_gap)

#left wg
l0_1 = x_pos1-x_pos0-length0
Taper = pg.taper(length = l0_1, width1 = wg_width_ic, width2 = wg_width_m2m, port = None, layer = 30)
D << Taper.move([x_pos0+length0,y_pos1])

#right coupler wg
rf1xl = li.rfim(wg_width_m2m, off_chip, W, L_tp, W_tp, L_mmi, W_mmi,Y_mmi,length1, radius,0,0,middle_e_width,e_e_gap)[1]
inter_euler_x =li.mod_euler(radius = interR, angle = interA)[1][0]
inter_euler_y =li.mod_euler(radius = interR, angle = interA)[1][1]
xposR = x_pos1+length1+2*rf1xl
Q = Path()
Q.append( pp.euler(radius = interR, angle = -interA) )
Q.append( pp.euler(radius = interR, angle = interA) )
Q.movex(xposR)
Q.movey(y_pos1)
X = CrossSection()
X.add(width = wg_width_m2m, offset = 0, layer = 30)
D << Q.extrude(cross_section = X)

Ramp = pg.ramp(length = coupler_lp, width1 = wg_width_m2m, width2 = wg_width_ic, layer = 30)
Ramp.mirror(p1 = [0,0], p2 = [1,0])
D << Ramp.move([xposR+2*inter_euler_x,y_pos1-2*inter_euler_y+wg_width_m2m/2])

D << li.fang(wg_width_ic,length=50*um,orientation='RU').move([xposR+2*inter_euler_x+coupler_lp,y_pos1-2*inter_euler_y-(wg_width_ic-wg_width_m2m)/2])
D.write_gds('Ising0.gds')
```

EOM RF2 (Main->Memory Cavity IM)

```python
length2 = 4*mm
couplergap = 50*um
fang_l = 90*um
x_pos2 = x_pos1+2*rf1xl+coupler_lall+coupler_lp+4*inter_euler_x+length1+couplergap+2*fang_l
y_pos2 = height/2
D << li.eom23(wg_width_m2m, off_chip, W, L_tp, W_tp, L_mmi, W_mmi,Y_mmi,length2, radius,x_pos2,y_pos2,middle_e_width,e_e_gap)

#left
P = Path()
P.append( pp.euler(radius = interR, angle = interA) )
P.append( pp.euler(radius = interR, angle = -interA) ) 
P.movex(x_pos2)
P.movey(y_pos2)
P.rotate(180, center = [x_pos2,y_pos2])
X = CrossSection()
X.add(width = wg_width_m2m, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

inter_euler_x =li.mod_euler(radius = interR, angle = interA)[1][0]
inter_euler_y =li.mod_euler(radius = interR, angle = interA)[1][1]

Ramp = pg.ramp(length = coupler_lall, width1 = wg_width_ic, width2 = wg_width_m2m, layer = 30)
Ramp.mirror(p1 = [0,0], p2 = [1,0])
D << Ramp.move([x_pos2-2*inter_euler_x-coupler_lall,y_pos2-2*inter_euler_y+wg_width_m2m/2])

D << li.fang(wg_width_ic,length=50*um,orientation='LU').move([x_pos2-2*inter_euler_x-coupler_lall,y_pos2-2*inter_euler_y-(wg_width_ic-wg_width_m2m)/2])


#right
X1 = CrossSection()#Create first CrossSection
X1.add(width = wg_width_m2m, layer = 30, name = 'wg2')
X2 = CrossSection() # Create second CrossSection to transition to
X2.add(width = wg_width_ic, layer = 30, name = 'wg2')
# Create the transitional CrossSection
Xtrans = pp.transition(cross_section1 = X1,cross_section2 = X2,width_type = 'sine')
# Create a Path for the transitional CrossSection to follow
Pc2 = Path()
Pc2.append( pp.euler(radius = interR, angle = interA) )
Pc2.append( pp.euler(radius = interR, angle = -interA) )
Pc2.movex(x_pos2+length2+2*rf1xl)
Pc2.movey(y_pos2)
# Use the transitional CrossSection to create a Device
WG_trans = Pc2.extrude(Xtrans)
D << WG_trans

Ramp = pg.ramp(length = coupler_lp, width1 = wg_width_ic, width2 = wg_width_m2m, layer = 30)
D << Ramp.move([x_pos2+length2+2*rf1xl+2*inter_euler_x,y_pos2+2*inter_euler_y-wg_width_ic/2])

x_ht = x_pos2+length2+2*rf1xl+2*inter_euler_x+coupler_lp
y_ht = y_pos2+2*inter_euler_y-(wg_width_ic-wg_width_m2m)/2
D << li.fang(wg_width_m2m,length=100*um,orientation='RD').move([x_ht,y_ht])
D.write_gds('Ising0.gds')
```

Heater Bottom

```python
hb_sx =xbb+2*rVg_euler_x
hb_sy = wg_y_off+g_heater+wg_width_oc/2
hb_ex =x_pos2-pad
D << pg.rectangle( size=(x_pos2-hb_sx-100, w_heater), layer=60 ).move( [hb_sx,hb_sy] )
D << pg.rectangle( size=(pad,pad), layer=60 ).move( [hb_sx,hb_sy] )
D << pg.rectangle( size=(pad,pad), layer=60 ).move( [hb_ex-100,hb_sy] )
D.write_gds('Ising0.gds')
```

```python
#EOM RF3 (Memory Cavity Flush)
length = 4*mm
x_posRF3 = x_pos2
y_posRF3 = wg_y_off+2*middle_e_width+e_e_gap+middle_e_width/2+e_e_gap
#D << li.dcimthin(wg_width_ic, off_chip, W, L_tp, W_tp, L_mmi, W_mmi,Y_mmi,length, radius,x_posRF3,y_posRF3,middle_e_width,e_e_gap)[0]
D << li.eom23(wg_width_m2m, off_chip, W, L_tp, W_tp, L_mmi, W_mmi,Y_mmi,length2, radius,x_posRF3,y_posRF3,middle_e_width,e_e_gap)

D.write_gds('Ising0.gds')
```

Bottom Racetrack

```python
x_pos = rt_start+rt_r
y_pos = wg_y_off+wg_gap+coupler_g1+wg_width_oc/2+wg_width_ic/2

#left hand curve
P = Path() 
P.append( pp.arc(radius = rt_r, angle = 180) )
P.movex(x_pos)
P.movey(y_pos)
P.mirror(p1 = [x_pos,1], p2 = [x_pos,3])
X = CrossSection()
X.add(width = wg_width_ic, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

Ramp = pg.ramp(length = coupler_l1, width1 = wg_width_ic, width2 = wg_width_oc, layer = 30)
Ramp.mirror(p1 = [0,0], p2 = [1,0])
D << Ramp.move([rt_start+rt_r,wg_gap+wg_y_off+wg_width_oc/2+coupler_g1+wg_width_ic])
D << li.fang(wg_width_oc,length=100*um,orientation='RU').move([x_pos+coupler_l1,y_pos+(wg_width_ic-wg_width_oc)/2])

## Top Curve
y_h = y_pos + 2*rt_r
l_rfeom0 = length0 + 2*inter_euler_x + coupler_lall
l_rfeom1 = length1 + 2*inter_euler_x + l0_1 + 2*rf1xl 
P = Path()
P.append( pp.straight(length = l_rfeom0 + l_rfeom1) )  
P.movex(x_pos)
P.movey(y_h)
X = CrossSection()
X.add(width = wg_width_ic, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

# first coupler
Ramp = pg.ramp(length = coupler_lp, width1 = wg_width_ic, width2 = wg_width_m2m, layer = 30)
D << Ramp.move([x_pos + l_rfeom0 + l_rfeom1,y_h-wg_width_ic/2])

# Inter coupler 1 and 2
x_gap = x_pos2-coupler_lall - 2*inter_euler_x -(x_pos + l_rfeom0 + l_rfeom1 + coupler_lp)
P = Path()
P.append( pp.straight(length = x_gap) )  
P.movex(x_pos + l_rfeom0 + l_rfeom1 + coupler_lp)
P.movey(y_h -(wg_width_ic-wg_width_m2m)/2)
X = CrossSection()
X.add(width = wg_width_m2m, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

# second coupler
Ramp = pg.ramp(length = coupler_lall, width1 = wg_width_m2m, width2 = wg_width_ic, layer = 30)
D << Ramp.move([x_pos + l_rfeom0 + l_rfeom1 + coupler_lp + x_gap,y_h-wg_width_ic/2])

# Past second coupler
l_rfeom2 = length2 + 4*inter_euler_x + 2*rf1xl 
xposrfeom2cr = x_pos + l_rfeom0 + l_rfeom1 + coupler_lp + x_gap+coupler_lall
P = Path()
P.append( pp.straight(length = l_rfeom2 + coupler_lp) ) 
P.movex(xposrfeom2cr)
P.movey(y_h)
X = CrossSection()
X.add(width = wg_width_ic, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

# QPM Region
YQPM_b = wg_gap+wg_y_off+(wg_width_ic-wg_width_oc)/2
target = (y_posRF3-YQPM_b)/2
rQb = 100*um
angleQb = li.find_ey(target,radius=rQb)
qpm_euler_x =li.mod_euler(radius = rQb, angle = angleQb)[1][0]
qpm_euler_y =li.mod_euler(radius = rQb, angle = angleQb)[1][1]
qpm_length = -coupler_l1+l_rfeom0 + l_rfeom1 + coupler_lp + x_gap
gap_rf3 = x_posRF3 - (qpm_length+rt_start+rt_r+coupler_l1+2*qpm_euler_x)

#left of RF EOM3
P = Path()
P.append( pp.straight(length = qpm_length) ) 
P.append( pp.euler(radius = rQb, angle = angleQb) ) 
P.append( pp.euler(radius = rQb, angle = -angleQb) ) 
P.append( pp.straight(length = gap_rf3) )
P.movex(rt_start+rt_r+coupler_l1)
P.movey(YQPM_b)
X = CrossSection()
X.add(width = wg_width_ic, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

#right of RF EOM3
l_rfeom3 = length2 + 2*rf1xl
delta_rf3 = 2*(inter_euler_x - qpm_euler_x)
xr_rfeom3 = rt_start+rt_r+coupler_l1+l_rfeom3+qpm_length+gap_rf3+2*qpm_euler_x
P = Path()
P.append( pp.euler(radius = rQb, angle = -angleQb) ) 
P.append( pp.euler(radius = rQb, angle = angleQb) )
P.append( pp.straight(length = delta_rf3) ) 
P.movex(xr_rfeom3)
P.movey(YQPM_b+2*qpm_euler_y)
X = CrossSection()
X.add(width = wg_width_ic, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

l_ramp_rf3 = xr_rfeom3+2*qpm_euler_x+delta_rf3
Ramp = pg.ramp(length = coupler_l1, width1 = wg_width_ic, width2 = wg_width_oc, layer = 30)
D << Ramp.move([l_ramp_rf3,YQPM_b-wg_width_ic/2])

rf3_bend = 1*mm
y_wpm_r = YQPM_b-(wg_width_ic-wg_width_oc)/2
y_delta_rf3 = wg_y_off+V_Groove_Spacing - y_wpm_r 
rbr = 100*um
anglebr = li.find_ey(y_delta_rf3/2,radius=rbr)
rb_euler_x =li.mod_euler(radius = rbr, angle = anglebr)[1][0]
qpm_r_end = chip_width - (l_ramp_rf3+coupler_l1+rf3_bend-2*rb_euler_x)+xmargin

P = Path()
P.append( pp.straight(length = rf3_bend) ) 
P.append( pp.euler(radius = rbr, angle = anglebr) ) 
P.append( pp.euler(radius = rbr, angle = -anglebr) )
P.append( pp.straight(length = qpm_r_end) )
P.movex(l_ramp_rf3+coupler_l1)
P.movey(y_wpm_r)
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

#right hand curve bottom
Ramp = pg.ramp(length = coupler_l1, width1 = wg_width_oc, width2 = wg_width_ic, layer = 30)
Ramp.mirror(p1 = [0,0], p2 = [1,0])
D << Ramp.move([l_ramp_rf3,YQPM_b+coupler_g1+wg_width_ic/2+wg_width_oc])
ybclf = YQPM_b+coupler_g1+wg_width_ic/2+wg_width_oc/2
D << li.fang(wg_width_oc,length=70*um,orientation='LU').move([l_ramp_rf3,ybclf])

ybc = y_posRF3 + 2*e_e_gap + 2.5*middle_e_width
ybcrf = ybclf - (wg_width_ic-wg_width_oc)/2
rbcr = (ybc - ybcrf)/2

P = Path()
P.append( pp.arc(radius = rbcr, angle = 180) ) 
P.movex(l_ramp_rf3+coupler_l1)
P.movey(ybcrf)
X = CrossSection()
X.add(width = wg_width_ic, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

#bend
Lbc = 13*mm
Δ = y_h - (ybcrf+2*rbcr)
ΔR = 20*um
R1 = (Δ-ΔR)/2

P = Path()

n_r = 3; 
r_sturn = (R1-ΔR *(n_r-1)-ΔR/2)/2
P.append( pp.straight(length = R1) ) 
for k in range(n_r):
    P.append( pp.straight(length = Lbc) )  
    P.append( pp.arc(radius = R1-ΔR*k, angle = 180) )

P.append( pp.straight(length = Lbc) )  
P.append( pp.arc(radius = r_sturn, angle = 180) )
P.append( pp.straight(length = Lbc) )  
P.append( pp.arc(radius = r_sturn, angle = -180) )

for k in range(n_r-1):
    P.append( pp.straight(length = Lbc) )
    P.append( pp.arc(radius = R1-ΔR*(n_r-1-k), angle = -180) )

P.append( pp.straight(length = Lbc+R1) ) 
P.append( pp.arc(radius = R1, angle = -180) )
P.append( pp.straight(length = l_ramp_rf3+coupler_l1-(xposrfeom2cr + l_rfeom2 + coupler_lp)) )
P.mirror(p1 = [0,0], p2 = [0,1])
P.movex(l_ramp_rf3+coupler_l1)
P.movey(ybcrf+2*rbcr)

X = CrossSection()
X.add(width = wg_width_ic, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

#DC EOM 4
elec_wbc = 30*um
e_e_gapbc = 10*um
bc_elec_x = l_ramp_rf3+coupler_l1-Lbc-R1
bc_elec_y = ybcrf+2*rbcr+3*ΔR+2*r_sturn
D << pg.rectangle( size=(Lbc, elec_wbc), layer=40 ).move([bc_elec_x,bc_elec_y+e_e_gapbc/2])
D << pg.rectangle( size=(Lbc, elec_wbc), layer=40 ).move([bc_elec_x,bc_elec_y-e_e_gapbc/2-elec_wbc])

D.write_gds('Ising0.gds')
```

```python
# spiral paramter 
Rspe = 100*um

# Left wg + ramp
xtrs = qpm_t_x+qpm_length_t+coupler_l1
ytrs = qpm_t_y+(wg_width_ic-wg_width_oc)/2

yd = (ytrs - (wg_y_off+6*V_Groove_Spacing))/2
rVgif = 100*um
anglerVgif = li.find_ey(yd,radius=rVgif)
rVg_euler_xif =li.mod_euler(radius = rVgif, angle = anglerVgif)[1][0] 

P = Path()
P.append( pp.straight(length = rt_r) )
P.append( pp.euler(radius = rVgif, angle = -anglerVgif) )
P.append( pp.euler(radius = rVgif, angle = anglerVgif) )
P.movex(xtrs)
P.movey(ytrs)
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

coupler_lp2 = 100*um
Ramp = pg.ramp(length = coupler_lp2, width1 = wg_width_oc, width2 = wg_width_nif, layer = 30)
Ramp.mirror(p1 = [0,0], p2 = [1,0])
D << Ramp.move([xtrs+2*rVg_euler_xif+rt_r, wg_y_off+6*V_Groove_Spacing+wg_width_oc/2])

#wg to output
x_wg2output = xtrs+2*rVg_euler_xif+rt_r+coupler_lp2
P = Path()
P.append( pp.straight(length = chip_width-x_wg2output+xmargin+coupler_lp2) )
P.movex(x_wg2output)
P.movey(wg_y_off+6*V_Groove_Spacing+(wg_width_oc-wg_width_nif)/2)
X = CrossSection()
X.add(width = wg_width_nif, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

#wg eomdc 6 first coupler bend
xramp_tb = xtrs+2*rVg_euler_xif+rt_r
yramp_tb = wg_y_off+6*V_Groove_Spacing-coupler_g3-wg_width_nif-wg_width_oc/2
coupler_lp2 = 100*um
Ramp = pg.ramp(length = coupler_lp2, width1 = wg_width_nif, width2 = wg_width_oc, layer = 30)
D << Ramp.move([xramp_tb, yramp_tb])
D << li.fang(wg_width_nif,length=100*um,orientation='LD').move([xramp_tb, yramp_tb+wg_width_nif/2])


rV6 = 100*um
yV6h = yramp_tb+(wg_width_oc)/2
h_target = yV6h - (wg_y_off+5*V_Groove_Spacing)
anglerV6 = li.find_ey(h_target/2,radius=rV6)
rV6_euler_x =li.mod_euler(radius = rV6, angle = anglerV6)[1][0] 
P = Path()
P.append( pp.euler(radius = rV6, angle = -anglerV6) )
P.append( pp.euler(radius = rV6, angle = anglerV6) )
P.movex(xramp_tb+coupler_lp2)
P.movey(yV6h)
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

MMIl = 2*L_tp+L_mmi
eom6_x = xramp_tb+coupler_lp2+2*rV6_euler_x+MMIl
eom6_y = wg_y_off+5*V_Groove_Spacing+Y_mmi/2
D << li.mmi(W, L_tp, W_tp, L_mmi, W_mmi, Y_mmi).move([xramp_tb+coupler_lp2+2*rV6_euler_x,wg_y_off+5*V_Groove_Spacing])

if_length = 2.2*mm
P = Path()
P.append( pp.straight(length = Rspe) )
P.movex(eom6_x)
P.movey(eom6_y)
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

DC3 = li.dcpm(if_length-2*Rspe,elec_w/2,e_e_gap,via,wg_width_oc)
D << DC3.move([eom6_x+Rspe,eom6_y])

P = Path()
P.append( pp.straight(length = Rspe) )
P.movex(eom6_x+Rspe+if_length-2*Rspe)
P.movey(eom6_y)
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

#right mmi
xr_mmi = xramp_tb+coupler_lp2+2*rV6_euler_x+if_length+2*MMIl
MMI_if_r = li.mmi(W, L_tp, W_tp, L_mmi, W_mmi, Y_mmi)
MMI_if_r.mirror(p1 = [0,0], p2 = [0,1])
D << MMI_if_r.move([xr_mmi,wg_y_off+5*V_Groove_Spacing])

#right wg
ifrwgl = chip_width-xr_mmi+xmargin+Y_mmi
P = Path()
P.append( pp.straight(length = ifrwgl) )
P.movex(xr_mmi)
P.movey(wg_y_off+5*V_Groove_Spacing)
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

#spiracl
diamter_sp = 4*V_Groove_Spacing-elec_w-e_e_gap
r_if = diamter_sp/6

num_turns = 6
in_gap = 200*um 
wg_wg_gap = (diamter_sp - in_gap)/(2*num_turns)
wg_wg_sep_margin = 10*um

y_discrepancy = 14.101

spiral_xs = xramp_tb+coupler_lp2+2*rV6_euler_x+MMIl
spiral_ys = wg_y_off+5*V_Groove_Spacing-Y_mmi/2
P = Path()
P.append( pp.arc(radius = Rspe, angle = -90) )
P.append( pp.straight(length = diamter_sp/2-Rspe) )
P.append( pp.spiral(num_turns=num_turns+1/2, gap=wg_wg_gap, inner_gap=in_gap, num_pts=10000) )
P.append( pp.straight(length = diamter_sp/2-Rspe) )
P.append( pp.arc(radius = Rspe, angle = 90) )
P.append( pp.straight(length = diamter_sp-2*Rspe) )
#P.append( pp.straight(length = if_length) )
#P.append( pp.arc(radius = r_if, angle = 180) )
#P.append( pp.straight(length = if_length-Rspe-r_if-wg_wg_sep_margin) )
#P.append( pp.arc(radius = r_if, angle = -180) )
#P.append( pp.straight(length = if_length-Rspe-r_if-wg_wg_sep_margin-2*r_if) )
#P.append( pp.arc(radius = r_if, angle = 90) )
#P.append( pp.arc(radius = r_if, angle = -90) )
P.movex(spiral_xs)
P.movey(spiral_ys)
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

DC3 = li.dcpm(if_length,elec_w/2,e_e_gap,via,wg_width_oc)
D << DC3.move([spiral_xs,spiral_ys-diamter_sp-y_discrepancy])

P = Path()
P.append( pp.arc(radius = r_if, angle = 180) )
P.movex(spiral_xs+if_length)
P.movey(spiral_ys-diamter_sp-y_discrepancy)
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

dc3ml = if_length-Rspe-r_if-wg_wg_sep_margin
dc3ml_x = spiral_xs+if_length-dc3ml
dc3ml_y = spiral_ys-diamter_sp-y_discrepancy+2*r_if
DC3 = li.dcpm(dc3ml,elec_w/2,e_e_gap,via,wg_width_oc)
D << DC3.move([dc3ml_x,dc3ml_y])

P = Path()
P.append( pp.arc(radius = r_if, angle = 180) )
P.mirror(p1 = [0,0], p2 = [0,1])
P.movex(dc3ml_x)
P.movey(dc3ml_y)
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

dc3tl = dc3ml-2*r_if
DC3 = li.dcpm(dc3tl,elec_w/2,e_e_gap,via,wg_width_oc)
D << DC3.move([dc3ml_x,dc3ml_y+2*r_if])

P = Path()
P.append( pp.arc(radius = r_if, angle = 90) )
P.append( pp.straight(length = y_discrepancy) )
P.append( pp.arc(radius = r_if, angle = -90) )
P.movex(dc3ml_x+dc3tl)
P.movey(dc3ml_y+2*r_if)
X = CrossSection()
X.add(width = wg_width_oc, offset = 0, layer = 30)
D << P.extrude(cross_section = X)

D.write_gds('Ising0.gds')
```
